```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV files into DataFrames
df1 = pd.read_csv('Smart_homes.csv')
df2 = pd.read_csv('Web_browsing.csv')
df3 = pd.read_csv('Online_gaming.csv')


# Select relevant columns for each dataset
df1_selected = df1[['Downlink_delay', 'Downlink_pcktloss', 'sta_nrg', 'Downlink_thrpt']]
df2_selected = df2[['Downlink_delay', 'Uplink_pcktloss', 'sta_nrg', 'Downlink_thrpt']]
df3_selected = df3[['Uplink_delay', 'Downlink_pcktloss', 'sta_nrg', 'Downlink_thrpt']]

# Rename columns for better presentation
df1_selected.rename(columns={
    'Downlink_delay': 'Latency (ms)',
    'Downlink_pcktloss': 'Packet Loss (%)',
    'sta_nrg': 'Energy Consumption (J)',
    'Downlink_thrpt': 'Throughput (Mbps)'
}, inplace=True)

df2_selected.rename(columns={
    'Downlink_delay': 'Latency (ms)',
    'Uplink_pcktloss': 'Packet Loss (%)',
    'sta_nrg': 'Energy Consumption (J)',
    'Downlink_thrpt': 'Throughput (Mbps)'
}, inplace=True)

df3_selected.rename(columns={
    'Uplink_delay': 'Latency (ms)',
    'Downlink_pcktloss': 'Packet Loss (%)',
    'sta_nrg': 'Energy Consumption (J)',
    'Downlink_thrpt': 'Throughput (Mbps)'
}, inplace=True)

# Add a new column to distinguish the datasets
df1_selected['Dataset'] = 'Smart Lighting'
df2_selected['Dataset'] = 'Web Browsing'
df3_selected['Dataset'] = 'Online Gaming'

# Combine the DataFrames
combined_df = pd.concat([df1_selected, df2_selected, df3_selected])

# Create the pairplot
sns.pairplot(combined_df, kind='scatter', hue='Dataset', palette='Set1', markers=["o", "s", "^"])

# Add a title
plt.suptitle('Demo scenarios', y=1.02)
plt.savefig("demos_multivariate_plot.png", bbox_inches = 'tight')
# Show the plot
plt.show()
```
